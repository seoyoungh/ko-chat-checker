```
def notNaN(x):
    return str(x) != str(1e400*0)
```

```
daily_chats = []
```

```
# 일상 대화 dataset의 내용들을 list로 만들기 
for i in range(len(daily)):
  d = np.asarray(daily)[i]
  for j in range(len(d)):
    if (notNaN(d[j])):
      daily_chats.append(d[j])
```

```
len(daily_chats)
```

```
# 단어 단위의 말뭉치 list 만들기
daily_word = []
for i in range(len(daily_chats)):
  temp_list = daily_chats[i].split()
  for temp in temp_list:
    daily_word.append(temp)
```

```
len(daily_word)
```

```
# 중복 제거
daily_words = list(set(daily_word))
len(daily_words)
```

```
(pd.DataFrame(daily_words)).to_csv("daily_word.csv")
```

```
# drama 1 에 있는 txt들 모두 모으기 
os.chdir("/content/gdrive/My Drive/Colab Notebooks/clb데이터분석캡스톤디자인/word data /일상 dataset/drama1")
import glob 
files = glob.glob('*.txt')
drama1 = []
for file in files:
    with open(file, 'r') as f:
        lines = f.readlines()
        for line in lines:
          drama1.append(line)
```

```
d1 = []
# 대화 dataset의 내용들을 list로 만들기 
for i in range(len(drama1)):
  line = drama1[i]
  for j in range(len(line)):
    if (notNaN(line[j])):
      d1.append(line[j])
# 단어 단위의 말뭉치 list 만들기
drama1_words = []
for i in range(len(d1)):
  if d1[i] == str(d1[i]):
    temp_list = d1[i].split()
  for temp in temp_list:
    drama1_words.append(temp)

len(drama1_words)
```

```
drama1_words = list(set(drama1_words))
print(len(drama1_words))
(pd.DataFrame(drama1_words)).to_csv("drama1_word.csv")
```

```
# drama 2 에 있는 txt들 모두 모으기 
os.chdir("/content/gdrive/My Drive/Colab Notebooks/clb데이터분석캡스톤디자인/word data /일상 dataset/drama2")
import glob 
files = glob.glob('*.txt')
drama2 = []
for file in files:
    with open(file, 'r') as f:
        lines = f.readlines()
        for line in lines:
          drama2.append(line)
```

```
d2 = []
# 대화 dataset의 내용들을 list로 만들기 
for i in range(len(drama2)):
  line = drama2[i]
  for j in range(len(line)):
    if (notNaN(line[j])):
      d2.append(line[j])
# 단어 단위의 말뭉치 list 만들기
drama2_words = []
for i in range(len(d2)):
  if d2[i] == str(d2[i]):
    temp_list = d2[i].split()
  for temp in temp_list:
    drama2_words.append(temp)

len(drama2_words)
```

```
drama2_words = list(set(drama2_words))
print(len(drama2_words))
(pd.DataFrame(drama2_words)).to_csv("drama2_word.csv")
```

```
# merge drama1 & drama2
drama_words = drama1_words + drama2_words
print(len(drama_words))
drama_words = list(set(drama_words))
print(len(drama_words))
(pd.DataFrame(drama_words)).to_csv("drama_word.csv")
```
